```python
# Create helper function for renaming
def rename_and_transform(df, year, mapping):
    """
    Renames columns and applies transformations to the dataset based on the year.
    
    Parameters:
        df (pd.DataFrame): The dataset to process.
        year (int): The year of the dataset.
        mapping (dict): The centralized column mapping.
    
    Returns:
        pd.DataFrame: The processed dataset.
    """
    # Rename columns
    if year in mapping:
        df = df.rename(columns=mapping[year])
        for old_col in mapping[year].keys():
            if old_col not in df.columns:
                print(f"Column '{old_col}' not found in {year} dataset. Skipping.")
    else:
        print(f"No column mapping found for year {year}. Skipping renaming.")

    # ✅ Fix GDP Transformation (Avoid Negative Log Values)
    if "gdp_per_capita" in df.columns:
        df["logged_gdp_per_capita"] = np.log1p(df["gdp_per_capita"] * 1000)  # Prevents log(0) error
        df = df.drop(columns=["gdp_per_capita"])
    elif "logged_gdp_per_capita" in df.columns:
        print(f"Logged GDP already exists in {year}, keeping it unchanged.")
    else:
        print(f"No GDP column found in {year} dataset. Skipping transformation.")

    # ✅ Fix Confidence Interval Calculation
    if "standard_error_of_happiness_score" in df.columns and "happiness_score" in df.columns:
        df["lower_confidence_interval"] = df["happiness_score"] - 2 * df["standard_error_of_happiness_score"]
        df["upper_confidence_interval"] = df["happiness_score"] + 2 * df["standard_error_of_happiness_score"]
    else:
        print(f"Required columns for confidence intervals not found in {year} dataset. Skipping transformation.")

    # ✅ Create 'happiness_rank' for Data Sets 2020-2023
    if "happiness_score" in df.columns and year in [2020, 2021, 2023]:
        df["happiness_rank"] = df["happiness_score"].rank(method='min', ascending=False)
    elif "happiness_score" not in df.columns:
        print(f"Column 'happiness_score' not found in {year} dataset. Skipping ranking.")

    return df
```

### 3. Data Cleaning 
#### g. Bring column values to one range format.

##### 2015-2019:
##### The healthy_life_expectancy values are likely normalized or scaled to a range between 0 and 1.
##### This is common in datasets where variables are standardized for comparison or modeling purposes.
##### 2020-2023:
##### The healthy_life_expectancy values are reported in actual years (e.g., 71.9, 72.4, etc.).
##### This represents the raw or unnormalized values of healthy life expectancy.
##### I choosed the option to bring normalize the 2020-2023 values to a range of 0 to 1 (or the same range as 2015-2019),
##### using the following formula:
##### Normalized Value = (Value − Min Value)/(Max Value-Min Value)


```python
# Reverse the 2015-2019 scaling using Min-Max transformation based on observed HLE real values from 2020-2023.
# Load the datasets
datasets = {
    2015: pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'world_happiness_2015_standardized.csv')),
    2016: pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'world_happiness_2016_standardized.csv')),
    2017: pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'world_happiness_2017_standardized.csv')),
    2018: pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'world_happiness_2018_standardized.csv')),
    2019: pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'world_happiness_2019_standardized.csv')),
    2020: pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'world_happiness_2020_standardized.csv')),
    2021: pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'world_happiness_2021_standardized.csv')),
    2022: pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'world_happiness_2022_standardized.csv')),
    2023: pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'world_happiness_2023_standardized.csv'))
}

# **Step 1: Get Min & Max Healthy Life Expectancy from 2020-2023 (real years)**
HLE_MIN = min(datasets[year]['healthy_life_expectancy'].min() for year in range(2020, 2024))
HLE_MAX = max(datasets[year]['healthy_life_expectancy'].max() for year in range(2020, 2024))

# **Step 2: Convert 2015-2019 Healthy Life Expectancy from (0-1) to (45-77)**
for year in range(2015, 2020):  # Only process 2015-2019
    if 'healthy_life_expectancy' in datasets[year].columns:
        print(f"Converting healthy_life_expectancy values for {year}...")

        datasets[year]['healthy_life_expectancy'] = (
            datasets[year]['healthy_life_expectancy'] * (HLE_MAX - HLE_MIN)
        ) + HLE_MIN

# **Step 3: Save ONLY the updated datasets for 2015-2019**
for year in range(2015, 2020):  # Only save changes for these years
    file_path = os.path.join(path, '02 Data', 'Prepared Data', f"world_happiness_{year}_standardized.csv")
    datasets[year].to_csv(file_path, index=False)
    print(f"✅ Saved updated dataset for {year}.")

```

```python
# Drop unnecessary columns
columns_to_drop = ['ladder_score_in_dystopia', 'explained_by:_log_gdp_per_capita', 'explained_by:_social_support', 
                   'explained_by:_healthy_life_expectancy', 'explained_by:_freedom_to_make_life_choices', 
                   'explained_by:_generosity', 'explained_by:_perceptions_of_corruption', 
                   'standard_error_of_ladder_score', 'explained_by:_gdp_per_capita', 'dystopia_residual', 'lower_confidence_interval', 'upper_confidence_interval', 'standard_error_of_happiness_score']
whr_cleaned = merged_report.drop(columns=columns_to_drop)

# Display the cleaned DataFrame
print(whr_cleaned)
```

```python
# Imputing missing values

# Sort by country and year
whr_cleaned.sort_values(by=['country', 'year'], inplace=True) 

# Forward fill and backward fill for 'region' column
whr_cleaned['region'] = whr_cleaned['region'].ffill()  # Forward fill
whr_cleaned['region'] = whr_cleaned['region'].bfill()  # Backward fill

# Impute numerical columns using country averages for 'happiness_score', 'social_support'
whr_cleaned['happiness_score'] = whr_cleaned.groupby('country')['happiness_score'].transform(lambda x: x.fillna(x.mean()))
whr_cleaned['social_support'] = whr_cleaned.groupby('country')['social_support'].transform(lambda x: x.fillna(x.mean()))
whr_cleaned['healthy_life_expectancy'] = whr_cleaned.groupby('country')['healthy_life_expectancy'].transform(lambda x: x.fillna(x.mean()))
whr_cleaned['freedom_to_make_life_choices'] = whr_cleaned.groupby('country')['freedom_to_make_life_choices'].transform(lambda x: x.fillna(x.mean()))
whr_cleaned['perceptions_of_corruption'] = whr_cleaned.groupby('country')['perceptions_of_corruption'].transform(lambda x: x.fillna(x.mean()))
whr_cleaned['logged_gdp_per_capita'] = whr_cleaned.groupby('country')['logged_gdp_per_capita'].transform(lambda x: x.fillna(x.mean()))
whr_cleaned['generosity'] = whr_cleaned.groupby('country')['generosity'].transform(lambda x: x.fillna(x.mean()))

# Check for remaining missing values
print(whr_cleaned.isnull().sum())
```

```python
# Impute 'happiness_rank' for datasets 2015-2019, 2021-2023
# Ensure the dataset is sorted by year and country
whr_cleaned.sort_values(by=['year', 'country'], inplace=True)

# Calculate 'happiness_rank' for each year
for year, df in whr_cleaned.groupby('year'):
    if 'happiness_score' in df.columns:
        # Calculate rank for the current year
        ranks = df['happiness_score'].rank(method='min', ascending=False)
        # Update the 'happiness_rank' column in the original dataset
        whr_cleaned.loc[df.index, 'happiness_rank'] = ranks
    else:
        print(f"Column 'happiness_score' not found in {year} dataset. Skipping ranking.")

# Check for remaining missing values
print(whr_cleaned.isnull().sum())    
```

### 4. Data Understanding 
#### a. Review Variables: Understand the meaning of each variable (e.g., GDP per capita, social support, etc.).

##### The original dataset brief lacked detailed explanations of the column meanings, which are crucial for conducting any analysis. By referring to the 2023 Global Happiness Report's (https://happiness-report.s3.amazonaws.com/2023/WHR+23_Statistical_Appendix.pdf) statistical appendix and the report itself (https://happiness-report.s3.amazonaws.com/2023/WHR+23.pdf), we gained a clearer understanding of the various columns.
##### •	The Happiness/Ladder Score represents the national average response to the question of life evaluations from the global survey, with the corresponding values of the standard error. All the bounds of the 95% confidence interval for the ladder score are also given.
##### •	GDP per capita is the Gross Domestic Product calculated in US Dollars, is used to determine the prosperity of countries based on their economic growth. GDP per capita is calculated by dividing the GDP of a nation by its population.
##### •	Healthy life expectancy(HLE) is a measure of the individual's life expectancy, in addition to taking their physical and mental health into account. HLE are based on the data extracted from the World Health Organization’s (WHO) Global Health Observatory data repository. 
##### •	Social support/Family is a measure of whether an individual have someone to count on in times of trouble (relatives or friends).
##### •	Freedom to make Life Choices represents the satisfaction of an individual's freedom to choose what to do with their life, which also includes their human rights.
##### •	Generosity reflects how generous an individual is, his/her sense to positive community engagement. For instance, whether he/she donated money to a charity in the past month.
##### •	Perception of Corruption is a measure of people's perceptions of corruption in their governments/business, and whether they have trust in their governments and the benevolence of others.
